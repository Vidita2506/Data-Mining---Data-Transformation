### Apply MinMaxScaler

```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn import preprocessing

wdata = load_wine(as_frame=True)

wdata.data.sample(10)
```

```python
minmax_scaler = preprocessing.MinMaxScaler()
wdata.data[list(wdata.data.columns)] = minmax_scaler.fit_transform(wdata.data[list(wdata.data.columns)])

wdata.data.sample(10)
```

```python
wdata.data.describe()
```

### Apply StandardScaler

```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn import preprocessing

wdata = load_wine(as_frame=True)

standard_scaler = preprocessing.StandardScaler()
wdata.data[list(wdata.data.columns)] = standard_scaler.fit_transform(wdata.data[list(wdata.data.columns)])

wdata.data.sample(10)
```

```python
wdata.data.describe()
```

### Apply PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)

X = wdata.data.iloc[:, 1:12].values
y = wdata.data.iloc[:, 12].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)


explained_variance = pca.explained_variance_ratio_
```
