```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

energy_df = pd.read_csv('./Energy_Consumption_by_Sector_monthly.csv')
energy_df[['Year','Month']] = energy_df.Month.str.split(" ",expand=True,)

data = energy_df.groupby(['Year'])[['Total Energy Consumed by the Residential Sector']].sum().round(2)
data = data.reset_index()

x = data['Year']
y = data['Total Energy Consumed by the Residential Sector']

xn = range(len(x))

plt.plot(xn, y)
plt.xticks(xn, x)

ax = plt.gca()
plt.xticks(rotation=90)

for index, label in enumerate(ax.xaxis.get_ticklabels()):
    if index % 5 != 0:
        label.set_visible(False)

plt.xlabel('Yearly')
plt.ylabel('Total Energy Consumed')
plt.title('Yearly energy consumption by the residential sector')
plt.show()
```
